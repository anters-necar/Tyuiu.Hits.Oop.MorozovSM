```polyglot-notebook
class Product{
    private string name;
    private decimal price;
    private int quantity;
    protected string category;
    internal string description{get;set;}
    public string Name{
        get {return name;}
        set {name = value;}
    }
    public decimal Price{
        get {return price;}
        set {price = value;}
    }
    public int Quantity{
        get {return quantity;}
        set {quantity = value;}
    }
    public string Category{
        get {return category;}
        set {category=value;}
    }

    public decimal GetTotalPrice(){
        return price*quantity;
    }
    public void UpdateQuantity(int amount){
        quantity += amount;
    }
    public string GetProductInfo(){
        return $"Товар: {name}, Цена: {price} руб., Количество: {quantity}, Категория: {category}, Описание: {description}";
    }
}
class Seller{
    private string name;
    private string employeeld;
    protected decimal salary;
    internal string contactInfo{get;set;}
    public List<Product> products{get;} = new List<Product>();

    public string Name{
        get {return name;}
        set {name = value;}
    }
    public string Employeeld{
        get {return employeeld;}
        set {employeeld = value;}
    }
    public decimal Salary{
        get {return salary;}
        set {salary = value;}
    }

    public void AddProduct(Product product){
        products.Add(product);
        Console.WriteLine($"Товар {product.Name} добавлен продавцу {name}");
    }
    public void SellProduct(Product product, int quantity){
        if (product.Quantity >= quantity)
        {
            product.UpdateQuantity(-quantity);
            Console.WriteLine($"Товар {product.Name} уменьшен на {quantity} единиц товара ");
        }
        else
        {
            Console.WriteLine($"Недостаточно товара {product.Name}");
        }
    }
    public string GetSellerInfo(){
        return $"Продавец: {name}, должность: {employeeld}, зарплата: {salary}, контактная информация{contactInfo}";
    }

}
class Store{
    private string storeName;
    private string location;
    public string storeHours{get;set;}
    private List<Seller> sellers = new List<Seller>();

    public string StoreName{
        get {return storeName;}
        set {storeName = value;}
    }
    public string Location{
        get {return location;}
        set {location = value;}
    }

    public void AddSeller(Seller seller){
        sellers.Add(seller);
        Console.WriteLine($"Продавец {seller.Name} добавлен магазину {storeName}");
    }
    public void ListProducts(){
        Console.WriteLine($"Товары в магазине {storeName}:");
        foreach (var seller in sellers){
            foreach (var product in seller.products)
            {
                Console.WriteLine(product.GetProductInfo());
            }
        }
    }
    public string GetStoreInfo()
    {
        return $"Магазин: {storeName}, Адрес: {location}, Часы работы: {storeHours}";
    }
}
Product product = new Product();
Seller seller = new Seller();
Store store = new Store();
product.Name = "Продукт";
product.Price = 10;
product.Quantity = 5;
product.Category = "Категория первая";
product.description = "Очень хороший продукт";
Console.WriteLine($"Общая стоимость всех единиц товара {product.GetTotalPrice()}");
seller.Name = "Иван";
seller.Employeeld = "Продавец";
seller.Salary = 1000;
seller.AddProduct(product);
seller.SellProduct(product,1);
seller.GetSellerInfo();
store.StoreName = "Магазин 1";
store.Location = "Первая улица";
store.storeHours = "8:00-20:00";
store.AddSeller(seller);
store.ListProducts();
store.GetStoreInfo();
```
